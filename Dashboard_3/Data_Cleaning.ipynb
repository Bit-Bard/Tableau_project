```python
replace_dict2={
    '"Major Market"': 'Major Market',
    '"Small Market"' :"Small Market",
}

combined_data['Market Size']=combined_data['Market Size'].replace(replace_dict2)

replace_dict3={
    '"Leaves"': "Leaves",
    '"Beans"' :"Beans",
}

combined_data['Product Line']=combined_data['Product Line'].replace(replace_dict3)

replace_dict4={
    '"Herbal Tea"': "Herbal Tea",
    '"Tea"' :"Tea",
    '"Espresso"':'Espresso',
    '"Coffee"' :'Coffee',  
}

combined_data['Product Type']=combined_data['Product Type'].replace(replace_dict4)

replace_dict5={
    '"Lemon"' : 'Lemon',
    '"Mint"' : "Mint",
    '"Darjeeling"': 'Darjeeling',
    '"Green Tea"' :'Green Tea',
    '"Decaf Espresso"':'Decaf Espresso',
    '"Decaf Irish Cream"':'Decaf Irish Cream',
    '"Amaretto"':'Amaretto',
    '"Colombian"':'Colombian',
    '"Caffe Mocha"': 'Caffe Mocha',
    '"Caffe Latte"':'Caffe Latte',
    '"Chamomile"':'Chamomile',
    '"Earl Grey"' : 'Earl Grey',
    '"Regular Espresso"':'Regular Espresso',
}

combined_data['Product']=combined_data['Product'].replace(replace_dict5)

replace_dict6={
    '"Colorado"':'Colorado',
    '"Texas"':'Texas',
    '"Florida"':'Florida',
    '"California"':'California',
    '"Iowa"':'Iowa',
    '"Connecticut"':'Connecticut',
    '"Oklahoma"':'Oklahoma',
    '"Nevada"' :'Nevada',
    '"Utah"':'Utah',
    '"New Hampshire"':'New Hampshire',
    '"Louisiana"':'Louisiana',
    '"Oregon"':'Oregon' ,
    '"Missouri"' :'Missouri',
    '"Wisconsin"' :'Wisconsin',
    '"Washington"':'Washington',
    '"Massachusetts"' : 'Massachusetts',
    '"Illinois"' :'Illinois',
    '"New Mexico"':'New Mexico',
    '"Ohio"':'Ohio',
    '"New York"':'New York'
}

combined_data['State']=combined_data['State'].replace(replace_dict6)

```
