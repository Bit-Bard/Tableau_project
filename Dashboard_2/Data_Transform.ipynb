```python
directory = 'archive (1)' 

columns_to_drop = ['uvIndex.1', 'moon_illumination', 'moonrise', 'moonset', 'totalSnow_cm', 'visibility', 'winddirDegree']

for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        file_path = os.path.join(directory, filename)

        # üè∑Ô∏è Extract city name robustly
        city_name = os.path.splitext(filename)[0]  # Remove .csv
        city_name = city_name.split('_')[-1].capitalize()  # Get 'Pune' from 'weather_pune.csv'

        print(f"‚úÖ Processing {filename} as city: {city_name}")  # Debug line

        df = pd.read_csv(file_path)
        df['City'] = city_name  # üí° Add city column here

        # Drop specified columns if they exist in the DataFrame
        df.drop(columns=[col for col in columns_to_drop if col in df.columns], inplace=True)

        for col in df:
            if pd.api.types.is_numeric_dtype(df[col]):
                df[col].fillna(df[col].mean(), inplace=True)

        # Save the cleaned DataFrame back to a CSV file
        cleaned_file_path = os.path.join(directory, f'{filename}')
        df.to_csv(cleaned_file_path, index=False)

        print(f"Cleaned data saved to {cleaned_file_path}")

print("All files have been processed.")
```

```python

directory = 'archive (1)'
dataframes = []

for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        
        file_path = os.path.join(directory, filename)
        df = pd.read_csv(file_path)

        dataframes.append(df)

concatenated_df = pd.concat(dataframes, ignore_index=True)

# Save the concatenated DataFrame to a new CSV file
concatenated_file_path = os.path.join(directory, 'Weather_Data_for_Indian_Cities.csv')
concatenated_df.to_csv(concatenated_file_path, index=False)

print(f"Concatenated data saved to {concatenated_file_path}")
```

```python
final=pd.read_csv("archive (1)/Weather_Data_for_Indian_Cities.csv")
final.head()
```

```python
columns_to_drop=['sunHour','uvIndex','DewPointC','FeelsLikeC','HeatIndexC','WindChillC','WindGustKmph','cloudcover','sunrise','sunset']

final.drop(columns=columns_to_drop, inplace=True)
final.head()
```

```python
final.shape
```

```python
final['date_time'] = pd.to_datetime(final['date_time'])
final['month'] = final['date_time'].dt.to_period('M')

# Group by City and Month
monthly_df = final.groupby(['City', 'month']).agg({
    'maxtempC': 'mean',
    'mintempC': 'mean',
    'tempC': 'mean',
    'humidity': 'mean',
    'precipMM': 'sum',
    'windspeedKmph': 'mean'
}).reset_index()

# Convert 'month' back to string format (e.g., 2023-06)
monthly_df['month'] = monthly_df['month'].astype(str)

# Save the final monthly summary
monthly_df.to_csv('monthly_weather_data_comp.csv', index=False)

print("‚úÖ Aggregated monthly weather data saved as 'monthly_weather_data_comp.csv'")
```
